```python
face = []
n_fc = []

for i in os.listdir('face detection/face'):
    face.append([cv.cvtColor(cv.imread('face detection/face/{}'.format(i), cv.IMREAD_COLOR), cv.COLOR_BGR2RGB) / 255, 1])

for i in os.listdir('face detection/non-face'):
    n_fc.append([cv.cvtColor(cv.imread('face detection/non-face/{}'.format(i), cv.IMREAD_COLOR), cv.COLOR_BGR2RGB) / 255, 0])

face = np.array(face)
n_fc = np.array(n_fc)

f_trn, f_tst = train_test_split(face, test_size = 0.2)
n_trn, n_tst = train_test_split(n_fc, test_size = 0.2)

train = []
test = []

for i in range(len(f_trn)):
    train.append(f_trn[i])
    train.append(n_trn[i])

for i in range(len(f_tst)):
    test.append(f_tst[i])
    test.append(n_tst[i])

random.shuffle(train)
random.shuffle(test)

train = np.array(train)
test = np.array(test)

trn_d = []
tst_d = []

trn_l = np.zeros(len(train))
tst_l = np.zeros(len(test))

for i in range(len(train)):
    trn_l[i] = train[i, 1]
    trn_d.append(train[i, 0])

for i in range(len(test)):
    tst_l[i] = test[i, 1]
    tst_d.append(test[i, 0])

trn_d = np.array(trn_d)
tst_d = np.array(tst_d)
```

```python
dtct_mdl = Sequential()

dtct_mdl.add(Conv2D(64, 3, activation = 'relu', input_shape = (48, 48, 3), padding = "same"))
dtct_mdl.add(MaxPooling2D(2, 2))

dtct_mdl.add(Conv2D(128, 3, activation = 'relu', padding = "same"))
dtct_mdl.add(MaxPooling2D(2, 2))

dtct_mdl.add(Conv2D(256, 3, activation = 'relu', padding = "same"))
dtct_mdl.add(Conv2D(512, 3, activation = 'relu', padding = "same"))

dtct_mdl.add(Flatten())
dtct_mdl.add(Dense(100, activation = 'relu'))
dtct_mdl.add(Dense(1, activation = 'sigmoid'))

dtct_mdl.compile(loss = 'binary_crossentropy', optimizer = 'adagrad', metrics = ['accuracy'])
```

```python
dtct_mdl.fit(trn_d, trn_l, epochs = 25, batch_size = 20, validation_data = (tst_d, tst_l))
```

```python
res = dtct_mdl.predict(tst_d)
ttl = 0

for i in range(len(tst_l)):
    if res[i] > 0.5 and tst_l[i] == 1:
        ttl += 1
    
    if res[i] <= 0.5 and tst_l[i] == 0:
        ttl += 1

print('Precision: ' + str(ttl / len(tst_l)))

matrix = metrics.confusion_matrix(tst_l, np.round(res))
disp = metrics.ConfusionMatrixDisplay(confusion_matrix = matrix, display_labels = list(set(tst_l)))
disp.plot()
plt.show()
```

```python
anlz_mdl = Sequential()

anlz_mdl.add(Conv2D(64, 3, activation = 'tanh', input_shape = (48, 48, 3), padding = "same"))
anlz_mdl.add(MaxPooling2D(2, 2))

anlz_mdl.add(Conv2D(128, 3, activation = 'tanh', padding = "same"))
anlz_mdl.add(MaxPooling2D(2, 2))

anlz_mdl.add(Conv2D(256, 3, activation = 'tanh', padding = "same"))

anlz_mdl.add(Flatten())
anlz_mdl.add(Dense(100, activation = 'sigmoid'))
anlz_mdl.add(Dense(7, activation = 'softmax'))

anlz_mdl.compile(loss = 'categorical_crossentropy', optimizer = 'Adagrad', metrics = ['CategoricalAccuracy'])
```

```python
anlz_mdl.fit(tr_dt, tr_lb, epochs = 20, batch_size = 21, validation_data = (ts_dt, ts_lb))
```
